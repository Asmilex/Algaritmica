```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from pyecharts import Bar
```

```python
from pyecharts import Bar
columns = ['Jan','Feb']
data1 = [2.0,4.9]
data2 = [2.6,5.9]
bar = Bar('Prueba')
bar.add('Cosa',columns,data1,mark_line=['average'],mark_point=['max','min'])
bar.add('Cosos',columns,data2,mark_line=['average'],mark_point=['max','min'])
bar.render()
bar
```
